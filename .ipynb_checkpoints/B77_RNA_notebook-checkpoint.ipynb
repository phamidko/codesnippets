## Method to look at relative gene expression such as MX1/HPRT1 and log scale transformation
The method loads excel file and aggregate all time-points based on excel worksheets. For example, Excel file (B75 RNA MX1 HPRT1 Ct.xlsx) contains 4 worksheets (Bleed 9-15, 9-30, 10-13 and 10-28). Each worksheet has a list of mouse id, HPRT1 Ct-values and MX1 Ct-values. 


```python
def graphByLogCt (cohort_excel_file, cohort_group_tbl):
    cohort_excel = pd.ExcelFile(cohort_excel_file)
    cohort_sheets = cohort_excel.sheet_names
    df_timepoints = pd.DataFrame()
    for i in cohort_sheets:
        df = pd.read_excel(cohort_excel,i)
        df['id'] = df['id'].astype(str)
        df = cohort_group_tbl.merge(df, on='id',how='left')

        df['time-point'] = i
        df['Ct ratio'] = df['MX1'].div(df['HPRT1'], axis ="index") # normalize by HPRT1
        df['Log Ct'] = -np.log(df['Ct ratio'])  # negative log transformation scale
        
        df_timepoints = df_timepoints.append(df)
        
    df_timepoints = df_timepoints.dropna(subset=['Log Ct'])
    df_timepoints = df_timepoints.reset_index(drop=True)
        
    plt.figure(figsize = (6, 6), dpi= 100)
    ax = sns.boxplot(x = "time-point", y = "Log Ct", data = df_timepoints, hue ='Group',
                     hue_order = ['Uninfected Control', 'Tumor only', 'HIV', 'HIV + Tumor', 'HIV + Tumor + ART'], showfliers=False, linewidth=1)
    
    # ----------------------------------------------------------------------------------------------------
    # prettify the plot

    # get the current figure
    ax = plt.gca()
    # get the xticks to iterate over
    xticks = ax.get_xticks()

    # iterate over every xtick and add a vertical line
    # to separate different classes
    #pos = range(len(nobs))

    for tick in xticks:
        ax.vlines(tick + 0.5, 0, np.max(df_timepoints["Log Ct"]), color = "grey", alpha = .1)
        #ax.text(pos[tick], medians[tick] + 0.03, nobs[tick],
                #horizontalalignment='center', size='small', color='black', weight='semibold')


    # rotate the x and y ticks
    ax.tick_params(axis = 'x', labelrotation = 90, labelsize = 12)
    ax.tick_params(axis = 'y', labelsize = 12)

    # add x and y label
    ax.set_xlabel("Post infection Time points", fontsize = 14)
    ax.set_ylabel("Relative MX1/HPRT1 expression", fontsize = 14)

    # add a title and put the legend on a specific location
    ax.set_title("Gene expression using Ct value", fontsize = 14)
    #ax.legend(loc = "lower left", fontsize = 'x-small');
    plt.legend(bbox_to_anchor=(1.35, 1),borderaxespad=0, fontsize = 'x-small')
    plt.show()
    return df_timepoints

```

## Normalize each timepoint using ΔΔCt and fold change
We normalize each timepoint Ct value with average mean value of uninfected control group.
First, we calculate average mean of MX1 and HPRT1 of uninfected control group; controlAvgRef_Mean and controlAvgTarget_Mean.
Next, substract that mean value from MX Ct which gives us ΔCt(MX1) of each mouse sample. Do same for reference gene expression ΔCt(HPRT1). 
To obtain ΔΔCt, all you have to do is substract ΔCt(MX1) from ΔCt(HPRT1). Lastly, we calculate using 2^-(ΔΔCt) and that is fold change value. All these calcualtion will be done in the following method, updateFoldChange


```python
def setBarcode(row):
    return row['time-point'] + ':' + row['id']

def setCt_ratio(row):
    return row['MX1'] / row['HPRT1'] 

def updateFoldChange (cohort_excel_file, cohort_group_tbl):
    cohort_excel = pd.ExcelFile(cohort_excel_file)
    cohort_sheets = cohort_excel.sheet_names
    df_timepoints = pd.DataFrame()
    for i in cohort_sheets:
        df = pd.read_excel(cohort_excel,i)
        df['id'] = df['id'].astype(str)
        df = cohort_group_tbl.merge(df, on='id',how='left')

        df['time-point'] = i
        df['Barcode'] = df.apply(setBarcode, axis =1)
        df['Ct ratio'] = df['MX1'].div(df['HPRT1'], axis ="index") # normalize by HPRT1
        df['Log Ct'] = -np.log(df['Ct ratio'])  # negative log transformation scale
        
        #df_control = pd.DataFrame()
        df_control = df.loc[df['Group'] == 'Uninfected Control']
        controlAvgRef_Mean = df_control['HPRT1'].mean()
        controlAvgTarget_Mean = df_control['MX1'].mean()
        
        # normailze delta Ct = MX1 of samples - average value MX1 of uninfected samples
        df['delta MX1'] ='0.0'
        df['delta HPRT1'] ='0.0'
        df['delta-delta Ct'] ='0.0'
        df['Fold Change'] ='0.0'
        df['delta MX1'] = df['delta MX1'].astype(float)
        df['delta HPRT1'] = df['delta HPRT1'].astype(float)
        df['delta-delta Ct'] = df['delta-delta Ct'].astype(float)
        df['Fold Change'] = df['Fold Change'].astype(float)
        
        df['delta MX1'] = df['MX1'] - controlAvgTarget_Mean        
        df['delta HPRT1'] = df['HPRT1'] - controlAvgRef_Mean
        df['delta-delta Ct'] = df['delta MX1'] - df['delta HPRT1']
        df['Fold Change'] = 2 ** -df['delta-delta Ct']

        #Aggregate each time-point and append to master data table
        df_timepoints = df_timepoints.append(df)
        
    #df_timepoints = df_timepoints.dropna(subset=['Log10 Ct'])
    df_timepoints = df_timepoints.reset_index(drop = True)
        
    return df_timepoints
```

## Graph B73 using normalized fold change

```python
%matplotlib inline
df_B73delta = pd.DataFrame()
df_B73delta = updateFoldChange(cohort_excel_file = 'B73 RNA MX1 HPRT1 Ct.xlsx', cohort_group_tbl = B73_group_tbl )


plt.figure(figsize = (5, 5), dpi= 100)
ax = sns.boxplot(x = "time-point", y = "Fold Change", data = df_B73delta, hue ='Group',
                     hue_order = ['Uninfected Control', 'Tumor only', 'HIV', 'HIV + Tumor', 'HIV + Tumor + ART'], showfliers=False, linewidth=1)
ax = plt.gca()
plt.legend(bbox_to_anchor=(1.35, 1),borderaxespad=0,fontsize = 'x-small' )
plt.title("B73 series - Normalized Fold Change Plot")
# rotate the x and y ticks
ax.tick_params(axis = 'x', labelrotation = 90, labelsize = 12)
ax.tick_params(axis = 'y', labelsize = 12)
```
